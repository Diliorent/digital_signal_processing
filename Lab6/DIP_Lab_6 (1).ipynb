1. Методами порогової та морфологічної обробки відобразити тільки контури циферблату

Завантажити зображення "digit.jpg" директорії "./examples/"


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/digit.jpg", 2)
kernel = np.ones((1,1),np.uint8)
opening = cv2.morphologyEx(img , cv2.MORPH_OPEN   , kernel, iterations = 1)
plt.imshow(opening, cmap = 'gray', vmin = 0, vmax = 255);
#plt.imshow(opening); plt.title('Відкриття');

plt.show();

ret, th = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
plt.figure(figsize=(20,20))
plt.subplot(236),plt.imshow(th, 'gray')
plt.show()
```
